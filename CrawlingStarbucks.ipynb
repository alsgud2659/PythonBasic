```python
from selenium import webdriver      #파이썬으로 브라우저 컨드롤 모듈
from bs4 import BeautifulSoup as bs # 크롤링한 데이터 정제
import pandas as pd
```

```python
driver = webdriver.Chrome('chromedriver.exe')
driver.get('https://www.starbucks.co.kr/store/store_map.do')
```

```python
area_btn = '#container > div > form > fieldset > div > section > article.find_store_cont > article > header.loca_search > h3 > a'
driver.find_element_by_css_selector(area_btn).click()
```

```python
seoul_bnt = "#container > div > form > fieldset > div > section > article.find_store_cont > article > article:nth-child(4) > div.loca_step1 > div.loca_step1_cont > ul > li:nth-child(1) > a"
driver.find_element_by_css_selector(seoul_bnt).click()
```

```python
all_bnt = '#mCSB_2_container > ul > li:nth-child(1) > a'
driver.find_element_by_css_selector(all_bnt).click()
```

```python
html = driver.page_source # 전체 페이지 소스 읽어오기
soup = bs(html)

 
```

```python
len(soup.select('ul.quickSearchResultBoxSidoGugun > li'))
```

```python
starbucks_list = soup.select('ul.quickSearchResultBoxSidoGugun > li.quickResultLstCon')
star = starbucks_list[1]
star
```

```python
# name = star["data-name"]
name = star.select('strong')[0].text.strip()
lat = star['data-lat'].strip()
lng = star['data-long'].strip()
addr = str(star.select('p')[0]).split('<br/>')[0].split('>')[1]
tel =  str(star.select('p')[0]).split('<br/>')[1].split('<')[0]
print(name)
print(lat)
print(lng)
print(addr)
print(tel)
```

```python
starbucks = []

for stars in starbucks_list:
    name = stars.select('strong')[0].text.strip()
    lat = stars['data-lat'].strip()
    lng = stars['data-long'].strip()
    addr = str(stars.select('p')[0]).split('<br/>')[0].split('>')[1]
    tel =  str(stars.select('p')[0]).split('<br/>')[1].split('<')[0]
    sigungu = str(stars.select('p')[0]).split('<br/>')[0].split('>')[1].split(' ')[1]
    
    
    starbucks.append([name, lat, lng, addr, sigungu, tel])

starbucks
```

```python
import pandas as pd

df = pd.DataFrame(starbucks, columns = ['위치', '위도', '경도', '주소','전화번호','시군구'])
df.to_excel('starbucks.xlsx', index = False)
df
```

```python
sgg_names = []

for addr in starbucks['주소']:
    sgg = addr.split()[1]
    sgg_names.append(sgg)
    
starbucks['시군구명'] = sgg_names

starbucks.head()
```

```python
starbucks_count = starbucks.pivot_table(
                                             index = '시군구명',
                                             values='매장명',
                                             aggfunc='count')
starbucks_count
```
